```
#Import libraries
from imutils import paths
import numpy as np
import imutils
import cv2
import os
```

```
print("[INFO] loading face detector...")
protoPath = "/content/drive/My Drive/Colab Notebooks/face_detection_model/deploy.prototxt"
modelPath = "/content/drive/My Drive/Colab Notebooks/face_detection_model/res10_300x300_ssd_iter_140000.caffemodel"
detector = cv2.dnn.readNetFromCaffe(protoPath, modelPath)
```

```
#Change  the dir name to crop every face
dir = "/content/drive/My Drive/Colab Notebooks/dataset1/test/aka" #change to test,train,val and for all students unk,pra,shr,rah,aka
a=os.listdir(dir)
a
for i in a:
    path =dir+"/"+i
    #read the image
    frame = cv2.imread(path)
    #Resize the image
    frame = imutils.resize(frame, width=800)
    (h, w) = frame.shape[:2]
    # construct a blob from the image
    imageBlob = cv2.dnn.blobFromImage(
        cv2.resize(frame, (200, 200)), 1.0, (200, 200),
        (104.0, 177.0, 123.0), swapRB=False, crop=False)

    # apply OpenCV's deep learning-based face detector to localize
    # faces in the input image
    detector.setInput(imageBlob)
    detections = detector.forward()
        # loop over the detections
    for i in range(0, detections.shape[2]):
        # extract the confidence (i.e., probability) associated with
        # the prediction
        confidence = detections[0, 0, i, 2]

        # filter out weak detections
        if confidence > 0.5:
            # compute the (x, y)-coordinates of the bounding box for
            # the face
            box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
            (startX, startY, endX, endY) = box.astype("int")

            # extract the face ROI
            face = frame[startY:endY, startX:endX]
            (fH, fW) = face.shape[:2]

            if fW < 40 or fH < 40:
                continue
            #save the croped image
            cv2.imwrite(path,face)  
                    
print("Done")
```
